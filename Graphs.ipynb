```python
#Retrieving CSV file
file_path = os.path.join("1. COVID_Reported_cases_per_month_ON.csv")
datafile_df = pd.read_csv(file_path)
datafile_df
```

```python
# Create the bar graph for the number of cases per month

Date = datafile_df["Date"]
Y = datafile_df["Number_of_Cases"]

# Creat the bar plot

plt.bar(Date,Y,color="lavender", width = 0.75,alpha = 1, align="edge")
plt.xticks(rotation = 40)
plt.tight_layout()
plt.title("Number of COVID cases reported in Ontario each month")
plt.xlabel("Month")
plt.ylabel("Number of cases reported")
plt.show()
```

```python
# Generate a line plot to show the reported cases flutuations

Date = datafile_df["Date"]
Y = datafile_df["Number_of_Cases"]

plt.plot(Date, Y)
plt.title("Fluctuation of the COVID cases reported in Ontario")
plt.xlabel("Date")
plt.ylabel("Number of Cases Reported")
plt.xticks(rotation = 40)
plt.show()
```

```python
response_two = quandl.get('CMHC/HPPA50_ON', start_date='2019-01-01', end_date='2020-09-30')
response_df = pd.DataFrame(response_two, columns = ['Average', 'Units']).reset_index()
response_df
```

```python
response_df['key_date'] = response_df['Date'].dt.strftime('%m/%Y')
response_df
```

```python
# Generate a line plot to show the house price flutuations

Date = response_df["Date"]
Y = response_df["Average"]

plt.plot(Date, Y)
plt.title("Fluctuation of the house price")
plt.xlabel("Date")
plt.ylabel("Average House Price (C$)")
plt.xticks(rotation = 40)
plt.show()
```

```python
# Create the bar graph for the number of houses sold per month

Date = response_df["key_date"]
Y = response_df["Units"]

# Creat the bar plot

plt.bar(Date,Y,color="lavender", width = 0.75,alpha = 1, align="edge")
plt.xticks(rotation = 40)
plt.tight_layout()
plt.title("Number of Houses Sold in Ontario each month")
plt.xlabel("Month")
plt.ylabel("Number of houses sold")
plt.show()
```
