```python
#Importing pandas & OS
import pandas as pd
import os
```

```python
#Retrieving CSV file
file_path = os.path.join("../CSV files/conposcovidloc.csv")
datafile_df = pd.read_csv(file_path)
#Changing case_reported_date to data format
datafile_df['Case_Reported_Date'] = pd.to_datetime(datafile_df['Case_Reported_Date'])
datafile_df.head()
```

```python
#Creating a column & extracting the month
datafile_df['month'] = pd.DatetimeIndex(datafile_df['Case_Reported_Date']).month
datafile_df.head()
```

```python
#Creating final dataframe with relevant info

COVID_ON_GroupBy_Month = datafile_df.groupby("month")
Reported_cases_per_month_ON = pd.DataFrame(COVID_ON_GroupBy_Month["month"].count())
Date = ["01/2020","02/2020","03/2020","04/2020","05/2020","06/2020","07/2020","08/2020","09/2020","10/2020","11/2020"]
Reported_cases_per_month_ON["Date"] = Date
Reported_cases_per_month_ON = Reported_cases_per_month_ON.reset_index(drop=True)
Reported_cases_per_month_ON.columns = ["Number_of_Cases","Date"]
Reported_cases_per_month_ON
```
