```python
import pandas as pd
import datetime as pt
import matplotlib 
```

```python
data_path = "Estimated_variable_mortgage_rate.csv"
interest_rate = pd.read_csv(data_path)
interest_rate
```

```python
interest_rate['Date_tmp'] = interest_rate["Date"]
#interest_rate.columns
interest_rate
```

```python
interest_rate['Date_tmp'] = pd.to_datetime(interest_rate['Date_tmp'])
interest_rate['Date_tmp'] = interest_rate['Date_tmp'].dt.strftime('%m/%Y')

interest_rate
```

```python
interest_rate_groupped_df = interest_rate.groupby(['Date_tmp'],as_index=False).mean()

interest_rate_groupped_df = interest_rate_groupped_df.rename(columns = {'Date_tmp':'Date'})

interest_rate_groupped_df
```
